# 📦 Live Tokenizacja Tekstu w Pythonie

Ten notebook pokazuje, jak działa tokenizacja tekstu przy użyciu tokenizerów OpenAI (`tiktoken`).

## 🔧 Import i przygotowanie tokenizera

```python
import tiktoken

# Pobieramy tokenizer używany w GPT-4-turbo
enc = tiktoken.encoding_for_model("gpt-4-turbo")
```

## ✨ Tokenizacja prostego tekstu

```python
text = "Examplary text to tokenize: 123 3445 000 :)"
tokens = enc.encode(text)

print(f"Tokeny: {tokens}")
print(f"Liczba tokenów: {len(tokens)}")
```

## 🔍 Wyświetlenie tokenów z dekodowaniem

```python
decoded_tokens = [enc.decode([token]) for token in tokens]

for idx, (token_id, token_text) in enumerate(zip(tokens, decoded_tokens)):
    print(f"Token {idx + 1}: ID={token_id}, Tekst='{token_text}'")
```

## 🎯 Nietypowe przykłady (emoji, polski tekst, liczby)

```python
examples = [
    "Hello 🌟 World!",
    "Programowanie w Pythonie 🐍",
    "1234567890",
    "Cześć! Jak się masz?",
    "AI is... unbelievable 😱🔥",
]

for example in examples:
    tokens = enc.encode(example)
    decoded_tokens = [enc.decode([token]) for token in tokens]
    print("\nTekst:", example)
    print("Tokeny:", decoded_tokens)
    print("Liczba tokenów:", len(tokens))
```
